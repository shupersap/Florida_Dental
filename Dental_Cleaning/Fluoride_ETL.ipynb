```python
import pandas as pd
```

```python
#obtaion all the data pertaining to the fluoride source
#only applicable years are 2019-2023, could have also used range, range(2019,2024)
years = [2019, 2020, 2021,2022, 2023]
#Initialize dictionary to hold dataframes for each year
FL_Fluoride = {} 
for year in years:
    df = pd.read_csv(f"../Clean_Data/Florida_Fluoride_{year}.csv", encoding='ISO-8859-1') # extract and read Fluoride csvs from sources folder
    df['year'] = year #label year column, for grouping if dfs are merged
    FL_Fluoride[f"{year}"] = df  
```

```python
FL_Fluoride["2019"]
```

```python
#Determine amount of null values present in the dataframe
for year in years:
    print("null count for" , year, "is:", FL_Fluoride[f"{year}"].isnull().sum().sum())
```

```python
merged_data = None
for year, df in FL_Fluoride.items():
# Add a year column to each DataFrame for merging purposes
    df["year"] = year    
#Merge into merged_data
    if merged_data is None:
        merged_data = df
    else:
        merged_data = pd.concat([merged_data, df], ignore_index=True)
```

```python
#save the merged file as a csv
merged_data.to_csv("../Clean_Data/Fluoride_2019-2023.csv",index=False)
```
