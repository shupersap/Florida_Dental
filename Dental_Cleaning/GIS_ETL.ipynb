```python
#function convert degrees and minute columns into common latitude and longitude
def ddmm_to_decimal(degrees, minutes):
    return degrees + (minutes / 60)
```

```python
#apply ddmm_to_decimal function for latitude and longitude
fl_gis['PLANT_LAT_DECIMAL'] = fl_gis.apply(lambda row: ddmm_to_decimal(row['PLANT_LAT_DD'], row['PLANT_LAT_MM']), axis=1)
fl_gis['PLANT_LONG_DECIMAL'] = fl_gis.apply(lambda row: ddmm_to_decimal(row['PLANT_LONG_DD'], row['PLANT_LONG_MM']), axis=1)
```

```python
#convert longitude coordinates to negative so it is recognized as west instead of east
fl_gis['PLANT_LONG_DECIMAL'] = -fl_gis['PLANT_LONG_DECIMAL']
```

```python
#only keep locating data
gis_subset= fl_gis[["PWS_ID","PLANT_LAT_DECIMAL","PLANT_LONG_DECIMAL"]]
```

```python
gis_subset=gis_subset.drop_duplicates(subset="PWS_ID")
```
